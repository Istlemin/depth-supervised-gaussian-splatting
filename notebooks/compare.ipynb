```python
from PIL import Image
import numpy as np
img_name = "00000"
baseline = "../output/chair/train/texture_alpha_10000/renders/"
method = "../output/chair_depth/train/texture_alpha_10000/renders/"

img_gt = np.array(Image.open(f"{baseline}{img_name}_gt.png")).astype(float)/255
img_baseline = np.array(Image.open(f"{baseline}{img_name}.png")).astype(float)/255
img_method = np.array(Image.open(f"{method}{img_name}_texture.png")).astype(float)/255

print(abs(img_gt-img_baseline).sum())
print(abs(img_gt-img_method).sum())
```

```python
import matplotlib.pyplot as plt
plt.imshow((abs(img_gt-img_baseline).sum(axis=2)-abs(img_gt-img_method).sum(axis=2))*10,vmin=-5,vmax=5,cmap="coolwarm")
plt.colorbar()
```

```python
method = "output/chair_diffuse_depth/train/texture_alpha_5000/renders/"

all_diff = []
for i in range(50):
    img_name = f"{i:0>5}"
    img_gt_depth = np.array(Image.open(f"{method}{img_name}gtdepth.png")).astype(float)/255
    img_depth = np.array(Image.open(f"{method}{img_name}_depth.png")).astype(float)/255
    all_diff.append(((img_gt_depth-img_depth*(img_gt_depth>0))**2).mean()/(img_gt_depth>0).mean())
    
print(np.mean(all_diff))
```
